```python
user_input = int(input('your number: '))

if user_input == 1:
    print('one')
elif user_input == 2:
    print('two')
else:
    print('many')
```

```python
user_input = int(input('your number: '))

if user_input == 1:
    print('one')
if user_input == 2:
    print('two')
if user_input not in [1, 2]:
    print('many')
```
