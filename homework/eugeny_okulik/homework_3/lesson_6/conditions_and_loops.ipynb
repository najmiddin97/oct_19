```python
user_input = int(input('your number: '))

if user_input == 1:
    print('one')
if user_input == 2:
    print('two')
if user_input not in [1, 2]:
    print('many')
```

# Loops - циклы

```python
names = ['John', 'Tom', 'James', 'Bob', 'Jim', 'Bill']

for name in names:
    print(name)
```

```python
names = ['John', 'Tim', 'James', 'Bob', 'Jim', 'Bill']

for name in names:
    name = name.replace('Tim', 'Texas')
    if name.startswith('J'):
        print('Mr.', end=' ')
    print(name)
```

Klyuch larini oldi

```python
persons = {'John': 132, 'Tom': 167, 'James': 234}
print(persons.keys())
for person in persons.keys():
    print(person) 

```

values - qiymatlarini chiqardi

```python
persons = {'John': 132, 'Tom': 167, 'James': 234}
print(persons.values())
for person in persons.values():
    print(person) 
```

Klyuch va znacheniyani oldi

```python
persons = {'John': 132, 'Tom': 167, 'James': 234}
for person in persons:
    print(f'{person}: {persons[person]}') 
```

Klyuch va znacheniyani oldi - optimizatsiya

```python
persons = {'John': 132, 'Tom': 167, 'James': 234}
print(persons.items())
for person in persons.items():
    #name, height = person
    print(f'{name}: {height}')
```

Optimizatsiya qisqartirildi

```python
persons = {'John': 132, 'Tom': 167, 'James': 234}
for name, height in persons.items():
    print(f'{name}: {height}')
```
