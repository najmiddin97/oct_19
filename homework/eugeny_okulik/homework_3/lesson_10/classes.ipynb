```python
from abc import abstractmethod

class Group:
    pupils = True
    school_name = 42
    director = 'Marivanna'

    def study(self):
        print('sit down and read')
    
    @abstractmethod
    def move(self):
        pass


class PrimaryGroup(Group):
    max_age = 11
    min_age = 6
    building_section = 'Left'

    def move(self):
        print('Run fast')


class HighGroup(Group):
    max_age = 16
    min_age = 12

    def move(self):
        print('go slowly')

        print('go')

```

Объект класса = объект типа = представитель класса = экземпляр клаксса

```python
first_a = PrimaryGroup()
eleven_a = HighGroup()

print(first_a.pupils)
print(first_a.min_age)
print(first_a.max_age)
print(first_a.building_section)
print(first_a.school_name)
first_a.study()
first_a.move()
eleven_a.move()
```
