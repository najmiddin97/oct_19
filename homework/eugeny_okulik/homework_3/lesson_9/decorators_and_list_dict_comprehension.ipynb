```python
def greet():

    def hello():
        return 'hello' 
    return hello()
print(greet())
```

```python
def outer():

    def inner():
        result = 2 + 5
        return result 
    return inner


inner_function = outer()
print(outer()())
print(inner_function())
```

```python
def func1(give_me_a_func): # "give_me_a_func" - bu yerga function beriladi qayerdan albatta pastdagi "simple" lardan
    print('before')  # before print qiladi
    give_me_a_func() # undan song yuqoridagi berilgan "simle" lardagi funksiya ishlaydi
    print('after') # after print qiladi


def simple1():
    print('simple1')

def simple2():
    print('simple2')

func1(simple1)
func1(simple2)

def simple3():
    print('I')
    print('need')
    print('automation')
    print('testing')
    print('course')

func1(simple3)
```

```python
def added_text(func):
    
    def wrapper():
        print('before')
        func()
        print('aftrer')
    
    return wrapper

def simple1():
    print('simple1')

simple1()

simple1 = added_text(simple1)

print(simple1)
simple1()



def simple2():
    print('simple2')

simple2 = added_text(simple2)
    
```

```python
def added_text(func):
    
    def wrapper():
        print('before')
        func()
        print('aftrer')
    
    return wrapper


@added_text
def simple1():
    print('simple1')

@added_text
def simple2():
    print('simple2')

simple1() # add_text(simple1)()
simple2()
```

```python
def add_logs(func):

    def wrapper():
        print(f'function {func.__name__} started')
        return func()
    
    return wrapper

@add_logs
def simple1():
    print('simple1')


@add_logs
def simple2():
    print('simple2')

simple1()
simple2()
```
