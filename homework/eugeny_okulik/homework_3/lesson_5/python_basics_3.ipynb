# Форматирование строки

```python
a = 'one'
b = 'two'
print('Firs word is', a, 'second word is', b)


my_text = 'Firs word is ' + a +', second word is '+ b
print(my_text)

my_text = 'First word is %s, second word is %s'
print(my_text % (a, b))

number1 = '1 orin'
number2 = '2 orin'

my_text2 = 'Real madrid %s, Barsa %s'
print(my_text2 %(number1, number2))

#string format
my_text = 'First word {0}, second word is {1}'
print(my_text.format(number1, number2))

#f- string
my_text = f'First word is {a}, second word is {b}'
print(my_text)

```
