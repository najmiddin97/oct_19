## Map

Приминение операции к каждому элементу

```python
my_list = [1, 2, 3, 4, 5]

new_list = []

for x in my_list:
    ## append(x * 2) - bu listdagi har bir elementni 2 ga kopaytirib chiqqan javobni new_listga beryapti
    new_list.append(x * 2)
print(new_list)

```

```python
my_list = [1, 2, 3, 4, 5]

def mult_bay_2(x):
    return x * 2

new_list = map(mult_bay_2, my_list)
print(list(new_list))
## map(mult_bay_2, my_list) map buyrugi my_list dagi har bir elementga Mult_bay_2 funksiyasini ishlatib chiqdi
```

```python
my_list = [1, 2, 3, 4, 5]

new_list = map(lambda x: x * 2, my_list)
print(list(new_list))
## lambda x: elementni qabul qiladi , u yordamida yuqoridagi def functionni qilgan ishini qildi
```

## Filter

```python
my_list2 = [1, 2, 3, 4, 5, 6, 7, 8]

new_list2 = []

for x in my_list2:
    if x % 2 == 0:
        new_list2.append(x)

print(new_list2)
## birinchi holat
```

```python
my_list2 = [1, 2, 3, 4, 5, 6, 7, 8]

def is_even(x):
    if x % 2 == 0:
        return True
    else:
        return False

new_list2 = filter(is_even, my_list2)
print(list(new_list2))
## optimiztasiya 1
```
